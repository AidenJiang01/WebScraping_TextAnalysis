# Text Analysis with Patent Abstract using Scikit Learn

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\user\Documents\RenewableEnergy_patents_2023_abstract.csv", index_col = 0)
```

```python
ab = df['abstract']
```

## Bag of Words: using min_df, stopwords, n-grams

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(min_df = 10, stop_words = "english", ngram_range = (1,3))
vect.fit(ab)
bof = vect.transform(ab)
print(len(vect.vocabulary_))
print(vect.vocabulary_.keys())
```

```python
print(repr(bof))
print(bof)
```

## tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer(min_df = 5, stop_words = "english", ngram_range = (1,3))
tfidf_vect.fit(ab)
tfidf_vect_bof = tfidf_vect.transform(ab)
print(len(tfidf_vect.vocabulary_))
print(tfidf_vect.vocabulary_.keys())
```

```python
print(tfidf_vect_bof)
```

```python
words_sortedby_tfidf = tfidf_vect_bof.toarray().max(axis = 0).argsort()
feature_names = tfidf_vect.vocabulary_.keys()

print('feautures with lowest tfidf:')
for i in list(words_sortedby_tfidf[:20]):
    print(f'{i}: ',list(feature_names)[i])

print('feautures with highest tfidf:')
for i in list(words_sortedby_tfidf[-20:]):
    print(f'{i}: ',list(feature_names)[i])
```

## Graph Representation of Bag of Words

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
dense_matrix = bof.toarray()

# token co-occurence matrix, token*token square matrix
token_matrix = np.dot(dense_matrix.T, dense_matrix)
np.fill_diagonal(token_matrix, 0)
```

```python
token_matrix.shape
```

```python
# create graph
G = nx.Graph()

# token list
tokens = list(vect.vocabulary_.keys())

# token as nodes
for token in tokens:
    G.add_node(token)

# add edges
for i in range(len(tokens)):
    for j in range(i + 1, len(tokens)):
        weight = token_matrix[i, j]
        if weight >= 10: ## filter out edges with weight lower than 10
            G.add_edge(tokens[i], tokens[j], weight=weight)
```

```python
# Define edge widths based on weights
edges = G.edges(data=True)
edge_widths = [data['weight'] for _, _, data in edges]
```

```python
# visualize token graph
plt.figure(figsize=(12, 12))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=True, node_size=80, node_color="gray", font_size=7, font_color = "blue", width = edge_widths/max(edge_widths))
plt.show()
```

## Graph Representation of Patents

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
dense_matrix = bof.toarray()

# patent concept overlapping matrix, patent*patent square matrix
patent_matrix = np.dot(dense_matrix, dense_matrix.T)
np.fill_diagonal(patent_matrix, 0)

patent_matrix.shape
```

```python
# create graph
G = nx.Graph()

# patent list
patents = list(df['id'])

# patent as nodes
for patent in patents:
    G.add_node(patent)

# add edges
for i in range(len(patents)):
    for j in range(i + 1, len(patents)):
        weight = patent_matrix[i, j]
        if weight >= 10: ## filter out edges with weight lower than 10
            G.add_edge(patents[i], patents[j], weight=weight)


# Define edge widths based on weights
edges = G.edges(data=True)
edge_widths = [data['weight'] for _, _, data in edges]
```

```python
# visualize token graph
plt.figure(figsize=(12, 12))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos, with_labels=False, node_size=20, node_color="grey", width = edge_widths/max(edge_widths))
plt.show()
```
